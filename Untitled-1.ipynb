```python
import heapq

class Graph:
    def __init__(self):
        self.adj_list = {}  # 邻接表存储：{顶点: [(邻接顶点, 权重), ...]}

    def add_vertex(self, v):
        if v not in self.adj_list:
            self.adj_list[v] = []

    def add_edge(self, v1, v2, weight):
        self.adj_list[v1].append((v2, weight))
        self.adj_list[v2].append((v1, weight))  # 无向图双向添加

    def dijkstra(self, start):
        # 初始化距离字典：key=顶点，value=距离（初始为无穷大）
        distances = {v: float('inf') for v in self.adj_list}
        distances[start] = 0
        # 优先队列：(当前距离, 顶点)，小顶堆自动排序
        heap = [(0, start)]
        visited = set()

        while heap:
            current_dist, current_v = heapq.heappop(heap)
            if current_v in visited:
                continue
            visited.add(current_v)

            # 遍历当前顶点的邻接顶点
            for neighbor, weight in self.adj_list[current_v]:
                distance = current_dist + weight
                # 如果新距离更短，更新并加入堆
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(heap, (distance, neighbor))
        return distances

# 测试
g = Graph()
g.add_vertex('A')
g.add_vertex('B')
g.add_vertex('C')
g.add_vertex('D')
g.add_vertex('E')
g.add_edge('A', 'B', 2)
g.add_edge('A', 'C', 4)
g.add_edge('B', 'E', 3)
g.add_edge('B', 'D', 1)
g.add_edge('C', 'D', 5)

print(g.dijkstra('A'))  # 输出：{'A':0, 'B':2, 'C':4, 'D':3, 'E':5}
```
